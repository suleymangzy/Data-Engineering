# Data Processing

Bu notebook'ta, R515B soğutucu akışkanına ait **Doymuş Buhar** ve **Kızgın Buhar** veri setlerindeki sütunların veri tipi dönüşümleri gerçekleştirilmektedir.

```python
import pandas as pd
import numpy as np
```

## Veri Setlerinin Yüklenmesi

Her iki veri setini Excel dosyalarından okuyoruz.

```python
df_doymus = pd.read_excel("R515B_Ozellikler/R515B_Doymus_Ozellikleri.xlsx")
df_kizgin = pd.read_excel("R515B_Ozellikler/R515B_Kizgin_Buhar.xlsx")
```

## Mevcut Veri Tiplerinin İncelenmesi

Her iki veri setindeki sütunların mevcut veri tiplerini inceliyoruz. Excel dosyalarında ondalık ayırıcı olarak virgül (`,`) kullanıldığından, tüm sütunlar `object` (metin) tipinde okunmuştur.

```python
df_doymus.dtypes
```

```python
df_kizgin.dtypes
```

## İlk Satırın Silinmesi

Her iki veri setindeki ilk satır, sütun açıklamalarını (girdi/çıktı bilgisi) içermekte olup gerçek veri değildir. Sayısal dönüşüme geçmeden önce bu satırları her iki DataFrame'den kaldırıyoruz.

```python
df_doymus = df_doymus.iloc[1:].reset_index(drop=True)
df_kizgin = df_kizgin.iloc[1:].reset_index(drop=True)

df_doymus.head()
```

```python
df_kizgin.head()
```

## Doymuş Buhar Özellikleri — Sayısal Tipe Dönüşüm

Doymuş buhar veri setindeki tüm sütunlar `object` (metin) tipinde okunmuştur. Ondalık ayırıcı olarak virgül (`,`) kullanıldığından, virgülleri noktaya (`.`) çevirerek tüm sütunları `pd.to_numeric` ile sayısal (`numeric`) tipe dönüştürüyoruz.

```python
df_doymus = df_doymus.apply(lambda col: pd.to_numeric(col.astype(str).str.replace(",", "."), errors="coerce"))
df_doymus.dtypes
```

```python
df_doymus.head()
```

## Kızgın Buhar Özellikleri — Sayısal Tipe Dönüşüm

Kızgın buhar veri setindeki sütunlar da aynı şekilde `object` tipinde okunmuştur. Virgülleri noktaya çevirerek sayısal tipe dönüştürüyoruz.

> **Not:** `state` sütunu kategorik/metin bilgisi içerdiğinden, sayısal dönüşüm sonrasında `NaN` değerler alacaktır. Bu sütun bir sonraki adımda veri setinden kaldırılacaktır.

```python
df_kizgin = df_kizgin.apply(lambda col: pd.to_numeric(col.astype(str).str.replace(",", "."), errors="coerce"))
df_kizgin.dtypes
```

```python
df_kizgin.head()
```

## `state` Sütununun Kaldırılması

Kızgın buhar veri setinde bulunan `state` sütunu, sayısal bir değer içermeyip kategorik/metin bilgisi taşımaktadır (örn. *superheated vapor*). Sayısal dönüşüm sonrası bu sütun tamamen `NaN` değerler içermektedir ve modelleme ile analiz süreçlerinde kullanılamaz. Bu nedenle veri setinden kaldırıyoruz.

```python
df_kizgin = df_kizgin.drop(columns=["state"])
df_kizgin.head()
```

## Dönüşüm Sonrası Doğrulama

Her iki veri setinin son durumunu kontrol ediyoruz. Tüm sütunların sayısal tipe dönüştüğünü, `state` sütununun kaldırıldığını ve dönüşüm sırasında oluşmuş olabilecek `NaN` değerleri doğruluyoruz.

### Eksik Değer Kontrolü

Dönüşüm sırasında `errors="coerce"` parametresi nedeniyle sayısal olmayan değerler `NaN`'a dönüştürülmüş olabilir. Her iki veri setindeki eksik değerleri kontrol ediyoruz.

```python
df_doymus.isnull().sum()
```

```python
df_kizgin.isnull().sum()
```
